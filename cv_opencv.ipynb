```python
import cv2
```

```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat_image.png")
cv2_imshow(image)
```

```python
print("SHAPE:", image.shape)
# 459 => height
# 431 => width
# 3   => color channels (RGB) Red, Green, Blue (BGR) Blue, Green, Red
print("HEIGHT:", image.shape[0])
print("WIDTH:", image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (300, 300))
cv2_imshow(resized)
```

```python
cropped = gray[100:300, 200:400]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped_cat.png", flipped)
```

```python
cv2.imwrite("cropped_cat.png", cropped)
```

# opencv video streaming codes** **bold text**

```python
print("Hello World!")

import cv2

cap = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "/haarcascade_frontalface_default.xml")

while True:
    success, frame = cap.read() # True, False | Boolean

    if not success:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("Real time face detection:", frame)



    if cv2.waitKey(1) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# ====================================================

# import cv2

# cap = cv2.VideoCapture(0)

# while True:
#     success, frame = cap.read() # True, False | Boolean

#     if not success:
#         break

#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     blurred = cv2.GaussianBlur(frame, (15, 15), 0) # Noice
#     edge1 = cv2.Canny(frame, 100, 200)
#     edge2 = cv2.Canny(frame, 50, 250)

#     flipped = cv2.flip(frame, 1)

#     # cv2.imshow("Original:", frame)
#     # cv2.imshow("gray:", gray)
#     # cv2.imshow("blurred:", blurred)
#     # cv2.imshow("edge1:", edge1)
#     # cv2.imshow("edge2:", edge2)
#     cv2.imshow("flipped:", flipped)




#     if cv2.waitKey(1) & 0xff == ord('q'):
#         break

# cap.release()
# cv2.destroyAllWindows()
```
